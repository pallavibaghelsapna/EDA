```python
#Time resampling
```

```python
df.head()
```

```python
df.resample(rule='A').min() # Here we can see every year last day is there and for the entire which was the minimum high ,minimum low,minimum open price and so on
#So this entire information is getting retrieved based on the minimum throughout the entire year
```

```python
df.resample(rule='A').max() # information is getting retrieved based on the maximum throughout the entire year
```

```python
## year and frequency
df.resample(rule='A').max()["OPEN"].plot() #Here we can see how open price decreases in one year
```

```python
##Bussiness end frequency
df.resample(rule='BA').max()['HIGH'].plot()
```

```python
df.resample(rule='BA').max()
```

```python
##plotting
df['OPEN'].resample(rule='BA').mean().plot(kind='bar')
```

```python
df['OPEN'].resample(rule='M').mean() # here is the monthly mean I am trying to get
```

```python
df['OPEN'].resample(rule='M').mean().plot(kind='bar') # Here we can see that what are the changes has been placed monthly
```

```python
df['HIGH'].rolling(11).mean().head(20) # we can find the rolling mean 
```

```python
df['HIGH'].rolling(11).max().head(20) # we can find the rolling max 
```

```python
df['open: 30 days rolling'] = df['OPEN'].rolling(30).mean()
```

```python
df.head(31)
```

```python
df[['OPEN','open: 30 days rolling']].plot(figsize=(12,5)) #here orange line is the smoothest version of this zigzag data
```
